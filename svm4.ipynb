```python
import pandas as pd
import numpy as np
data = pd.read_csv("hr_attrition.txt",sep="\t",header=0)
```

```python
data.shape
```

```python
data.head()
```

```python
data.loc[data.Attrition == 'Yes', 'Attrition'] = 1
data.loc[data.Attrition == 'No', 'Attrition'] = 0
data.head()
```

```python
data.dtypes
```

```python
data.BusinessTravel.unique()
```

```python
data.loc[data.BusinessTravel == 'Travel_Rarely', 'BusinessTravel'] = 1
data.loc[data.BusinessTravel == 'Travel_Frequently', 'BusinessTravel'] = 2
data.loc[data.BusinessTravel == 'Non-Travel', 'BusinessTravel'] = 0
data.Gender.unique()
data.loc[data.Gender == 'Male', 'Gender'] = 1
data.loc[data.Gender == 'Female', 'Gender'] = 0
```

```python
data = data.drop(columns=['Department','EducationField','JobRole','MaritalStatus','Over18','OverTime'])
```

```python
data.dtypes
```

```python
X = data.drop('Attrition', axis=1)
y = data['Attrition']
```

```python
X.head()
```

```python
y.head()
```

```python
print(X.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20)
```

```python
from sklearn.svm import SVC
svclassifier = SVC(kernel='linear')
svclassifier.fit(X_train, y_train)
```

```python
y_pred = svclassifier.predict(X_test)
```

```python
print(X_test.shape)
print(y_pred.shape)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
