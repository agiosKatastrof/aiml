```python
diabCodes = {}
with open("codes.txt") as f:
    for line in f:
        line = line.rstrip()
        (key,val) = line.split(" = ")
        diabCodes[int(key)] = val
```

```python
diabCodes
```

```python
data["CodeName"] = data["Code"].map(diabCodes)
```
