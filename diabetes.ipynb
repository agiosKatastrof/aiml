```python
import pandas as pd
import numpy as np
data = pd.read_csv("data-all-diabetes",sep="\t", names=["date","time","Code","Value"])
```

```python
diabCodes = {}
with open("codes.txt") as f:
    for line in f:
        line = line.rstrip()
        (key,val) = line.split(" = ")
        diabCodes[int(key)] = val
```

```python
diabCodes
```

```python
data["CodeName"] = data["Code"].map(diabCodes)
```

```python
data
```

```python
data['CodeName'].value_counts()
```

```python
print(data.dtypes)
print(data.applymap(np.isreal))
data.Value = pd.to_numeric(data.Value, errors='coerce')
print(data.dtypes)
print(data.applymap(np.isreal))
data.groupby(['CodeName']).mean()
```

```python
data = data.drop(data[data.Value == 0].index)
```
